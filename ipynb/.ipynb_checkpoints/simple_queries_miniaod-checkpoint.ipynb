## Build / Play with dimuon pairs

```python
from pyspark.sql import Row
import itertools
from math import *

# filter out muons that we do not need
# pt>10 and |eta| < 2.4
def filterMuons(event):
    muons = []
    fops = lambda muon: muon.m_state.p4Polar_.fCoordinates.fPt>10 and abs(muon.m_state.p4Polar_.fCoordinates.fEta)<2.4
    return Row(cmsMuons=filter(fops, event.cmsMuons))

def buildDimuon(muon1, muon2):
    pt1 = abs(muon1.m_state.p4Polar_.fCoordinates.fPt)
    phi1 = muon1.m_state.p4Polar_.fCoordinates.fPhi
    theta1 = 2.0*atan(exp(-muon1.m_state.p4Polar_.fCoordinates.fEta))
    px1 = pt1 * cos(phi1)
    py1 = pt1 * sin(phi1)
    pz1 = pt1 / tan(theta1)
    E1 = sqrt(px1**2 + py1**2 + pz1**2 + 0.10565836727619171**2)
    #
    pt2 = abs(muon2.m_state.p4Polar_.fCoordinates.fPt)
    phi2 = muon2.m_state.p4Polar_.fCoordinates.fPhi
    theta2 = 2.0*atan(exp(-muon2.m_state.p4Polar_.fCoordinates.fEta))
    px2 = pt2 * cos(phi2)
    py2 = pt2 * sin(phi2)
    pz2 = pt2 / tan(theta2)
    E2 = sqrt(px2**2 + py2**2 + pz2**2 + 0.10565836727619171**2)
    return sqrt((E1 + E2)**2 - (px1 + px2)**2 - (py1 + py2)**2 - (pz1 + pz2)**2)

# build dimuons
def buildDimuons(event):
#    return Row(first=event.cmsMuons[0].m_state.p4Polar_.fCoordinates.fPt, second=event.cmsMuons[1].m_state.p4Polar_.fCoordinates.fPt)
#    fops = lambda pair: pair[0].m_state.pdgId_ != pair[1].m_state.pdgId_
    dimuons = itertools.combinations(event.cmsMuons, 2)
    return itertools.imap(lambda pair: Row(mass=buildDimuon(pair[0], pair[1])), dimuons)
```

```python
masses = muons.rdd\
    .map(filterMuons)\
    .filter(lambda event: len(event.cmsMuons)>=2)\
    .flatMap(buildDimuons)
```

```python
from pyspark.sql import Row
import itertools
from math import *

# CMSSW reco::Muon definitions
GLOBAL_MUON = 1 << 1
TRACKER_MUON = 1 << 2
STANDALONE_MUON = 1 << 3
PFMUON = 1<<5

# true if number of valid vertices is > 0
def validVertex(row):
    return len(itertools.ifilter(
        lambda vertex: vertex.validity_ and abs(vertex.position_.fCoordinates.fZ)<24 and vertex.ndof_>4, 
        row.cmsVertices)) > 0

def isGlobalMuon(muon):
    return muon.type_ & GLOBAL_MUON

def isTrackerMuon(muon):
    return muon.type_ & TRACKER_MUON

def isPFMuon(muon):
    retur muon.type_ & PFMUON

def isLooseMuon(muon):
    return isPFMuoun(muon) and (isGlobalMuon(muon) || isTrackerMuon(muon))

def isMediumMuon(muon):
    return isLooseMuon(muon) and muon.combinedQuality_.chi2LocalPosition < 12 and muon.combinedQuality_.trkKink<20

def muonIsolation(muon):
    pfIso = muon.pfIsolationR04_
    return (pfIso.sumChargedHadronPt 
            + max(0., pfIso.sumNeutralHadronEt + pfIso.sumPhotonEt - 0.5*pfIso.sumPUPt)) / muon.m_state.p4Polar_.fCoordinates.fPt

def mapMuons(event):
    fops = lambda muon: muon.m_state.p4Polar_.fCoordinates.fPt>10\
        and abs(muon.m_state.p4Polar_.fCoordinates.fEta)<2.4 and isMediumMuon(muon) and\
        and isGlobalMuon(muon) and isTrackerMuon(muon)
        and muonIsolation(muon)<0.25
    return Row(cmsMuons=filter(fops, event.cmsMuons))

def buildDimuon(muon1, muon2):
    pt1 = abs(muon1.m_state.p4Polar_.fCoordinates.fPt)
    phi1 = muon1.m_state.p4Polar_.fCoordinates.fPhi
    theta1 = 2.0*atan(exp(-muon1.m_state.p4Polar_.fCoordinates.fEta))
    px1 = pt1 * cos(phi1)
    py1 = pt1 * sin(phi1)
    pz1 = pt1 / tan(theta1)
    E1 = sqrt(px1**2 + py1**2 + pz1**2 + 0.10565836727619171**2)
    #
    pt2 = abs(muon2.m_state.p4Polar_.fCoordinates.fPt)
    phi2 = muon2.m_state.p4Polar_.fCoordinates.fPhi
    theta2 = 2.0*atan(exp(-muon2.m_state.p4Polar_.fCoordinates.fEta))
    px2 = pt2 * cos(phi2)
    py2 = pt2 * sin(phi2)
    pz2 = pt2 / tan(theta2)
    E2 = sqrt(px2**2 + py2**2 + pz2**2 + 0.10565836727619171**2)
    return sqrt((E1 + E2)**2 - (px1 + px2)**2 - (py1 + py2)**2 - (pz1 + pz2)**2)

# build dimuons
def buildDimuons(event):
#    return Row(first=event.cmsMuons[0].m_state.p4Polar_.fCoordinates.fPt, second=event.cmsMuons[1].m_state.p4Polar_.fCoordinates.fPt)
#    fops = lambda pair: pair[0].m_state.pdgId_ != pair[1].m_state.pdgId_
    dimuons = itertools.combinations(event.cmsMuons, 2)
    return itertools.imap(lambda pair: Row(mass=buildDimuon(pair[0], pair[1])), dimuons)
```
